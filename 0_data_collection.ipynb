```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
def get_record(soup):
    title = soup.find('h1').get_text() 
    location = soup.find('div', class_='z1s6m00 _1hbhsw65a').get_text() 
    jd = soup.find('div', class_='_1x1c7ng0').get_text()  
    info = soup.find_all('div', class_ = 'z1s6m00 _5135ge0 _5135ge2')
    for i in info:
        if i.get_text().startswith('Additional Information'):
                try:
                    career_level = i.find_all('span')[1].get_text()
                except IndexError:
                    career_level = 'null'    
                try:
                    experience = i.find_all('span')[5].get_text()
                except IndexError:
                    experience = 'null'  
                try:
                    job_type = i.find_all('span')[7].get_text() 
                except IndexError:
                    job_type = 'null'  
    record = {'job_title': title,
              'location': location,
              'job_description': jd,
              'career_level': career_level,
              'experience': experience,
              'job_type' :job_type}
    return record
```

```python
def extract_job(u):
    r = requests.get(u)
    soup = bs(r.content, 'html.parser') # Parsing the HTML
    records = pd.DataFrame()
    #find all job-links embeded in one page
    for link in soup.find_all('a'):
        l = link.get('href')
        if l.startswith('/en/job/'):
            job_link = 'https://www.jobstreet.com.my' + l 
            
             #find job information in each link
            r = requests.get(job_link)
            soup = bs(r.content, 'html.parser')
            records = records.append(get_record(soup), ignore_index=True)
    return records   
```

```python
def extract_pages(page_from, page_to):      
    url = 'https://www.jobstreet.com.my/data-jobs/in-Malaysia?createdAt=7d&pg=1'
    records = extract_job(url)
    for i in range(page_from+1, page_to+1):
        url = 'https://www.jobstreet.com.my/data-jobs/in-Malaysia?createdAt=7d&pg=' + str(i)
        next_records = extract_job(url)
        records = pd.concat([records, next_records])
    return(records)
```

```python
df = extract_pages(1, 110)
```
