```python
import pandas as pd
from collections import Counter
from pprint import pprint
import matplotlib.pyplot as plt
```

### Load dataset

```python
skillset = pd.read_csv('skillset groups.csv')
print(skillset.groupby('kmeans')['skillset'].count())
skill_list = pd.DataFrame(skillset.groupby('kmeans')['skillset'].apply(list))
most_common= pd.DataFrame(columns=['group','most common skills','count'])
for i in skill_list.index:
    occurence = Counter(skill_list.iloc[i][0])
    most = occurence.most_common(10)
    for ele in most:
        new_row = {'group': i+1,'most common skills':ele[0],'count':ele[1]}
        most_common = pd.concat([most_common, pd.DataFrame([new_row])], ignore_index=True)
#most_common.to_csv('top_10_skills.csv', index=False)
```

```python
for i in skill_list.index:
    occurence = Counter(skill_list.iloc[i][0])
    most_common = occurence.most_common(30)
    most_common_skill = list(zip(*most_common))[0]
    pprint(most_common_skill)
```

```python
df = pd.read_csv('Data/4-clean data.csv')
df.fillna("", inplace=True)
print(df.shape)
df['skill_group1'] = 0
df['skill_group2'] = 0
df['skill_group3'] = 0
df['skill_group4'] = 0
df['skill_group5'] = 0
df['skill_group6'] = 0
df.head(3)
```

```python
for i in df.index:
    string = df["job_description"][i]
    for j in skill_list.index:
        x = skill_list['skillset'][j]
        if any(ele in string for ele in x):
            if j==0:
                df.loc[i, 'skill_group1'] = 1
            if j==1:
                df.loc[i, 'skill_group2'] = 1 
            if j==2:
                df.loc[i, 'skill_group3'] = 1  
            if j==3:
                df.loc[i, 'skill_group4'] = 1  
            if j==4:
                df.loc[i, 'skill_group5'] = 1   
            if j==5:
                df.loc[i, 'skill_group6'] = 1  
```

```python
df = pd.read_csv('Data/4-clean data.csv')
df.fillna("", inplace=True)
print(df.shape)
df['skill_group1'] = 0
df['skill_group2'] = 0
df['skill_group3'] = 0
df['skill_group4'] = 0
df['skill_group5'] = 0
df['skill_group6'] = 0
df.head(3)
```

```python
for i in df.index:
    string = df["job_description"][i]
    for j in skill_list.index:
        x = skill_list['skillset'][j]
        if any(len(ele) >= 8 and ele in string for ele in x):
            if j==0:
                df.loc[i, 'skill_group1'] = 1
            if j==1:
                df.loc[i, 'skill_group2'] = 1 
            if j==2:
                df.loc[i, 'skill_group3'] = 1  
            if j==3:
                df.loc[i, 'skill_group4'] = 1  
            if j==4:
                df.loc[i, 'skill_group5'] = 1   
            if j==5:
                df.loc[i, 'skill_group6'] = 1  
```

```python
x = 230
colors = ["#B9DDF1", '#006ea1']
for i in range(1,7):
    plt.subplot(x+i)
    t = 'Cluster ' + (str(i))
    plt.title(t)
    name = "skill_group" + (str(i))
    plt.pie(df[name].value_counts(),autopct='%1.1f%%',
           colors=[colors[v] for v in df[name].value_counts().keys()])
    plt.legend(df[name].value_counts().keys(), loc='upper right')
plt.show()
```
